```python
import sys
sys.path.append('../src')
from pypeal.entities.peal import Peal


peals = Peal.get_all()

len(peals)
```

```python
def peal_to_dict(peal: Peal):
    return {
        'id': peal.id,
        'date': peal.date,
        'title': peal.title,
        'location': peal.location,
        'location_detail': peal.location_detail,
        'duration': peal.duration,
        'composer': peal.composer,
        'conductors': peal.conductors,
        'length_type': peal.length_type,
        'changes': peal.changes,
        'ringers': [r.ringer for r in peal.ringers],
    }
```

```python
df = pd.DataFrame([peal_to_dict(peal) for peal in peals])

df
```

```python
df[['location_detail', 'location']].value_counts()
```

```python
# count peals that contain a specific ringer
ringer = 'Michael J Stubbs'
df['ringers'].apply(lambda ringers: ringer in [r.name for r in ringers]).value_counts()
```

```python
df_exploded = df.explode('ringers')
df_exploded['Ringer'] = df_exploded['ringers'].apply(lambda r: r.name)
df_exploded['Bell'] = df_exploded['ringers'].apply(lambda r: r.bell_nums and r.bell_nums[0])

df_exploded
```

```python
ringers_view = df['ringers'].explode().value_counts()

ringers_view
```

```python
ringers_view = df.explode('ringers')
ringers_view['name'] = ringers_view['ringers'].apply(lambda r: r.name)
ringers_view['given_names'] = ringers_view['ringers'].apply(lambda r: r.given_names)
ringers_view['last_name'] = ringers_view['ringers'].apply(lambda r: r.last_name)
# ringers_view.drop('ringers', axis=1)

ringers_view
```
